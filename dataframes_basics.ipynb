```python
import pandas as pd
import numpy as np
```

```python
#a dataframe is a collection of series with the same indexing

#dataframes can be made by setting columns 
dataframe_c = pd.DataFrame({'scores': [5, 6, 8], 'possible': [5, 7, 9]})
print(dataframe_c)

#dataframs can be made by combining lists
dataframe_l = pd.DataFrame(
[   [5, 5],
    [6, 7],
    [8, 9],],
    columns= ['scores', 'possible']
)

print(dataframe_l)

#.info() gives all info, .shape gives rows, columns, .axes gives row and column index, .size gives rows * columns
print(dataframe_c.info())
print(dataframe_c.shape)
print(dataframe_c.axes)
print(dataframe_c.size)

#use .rename() to change the name of columns
dataframe_c.rename(columns = {'scores' : 'real_scroes', 'possible' : 'max'}, inplace=1)

print(dataframe_c)
```

```python
#dataframes are very imortant when working with csv files

original_dataframe = pd.DataFrame({
    'sin': np.sin(np.linspace(-np.pi, np.pi, 100)),
    'tan': np.tan(np.linspace(-np.pi, np.pi, 100)),
    'cos' : np.cos(np.linspace(-np.pi, np.pi, 100))
    }
)

#to paste into a csv
original_dataframe.to_csv('trig.csv')

#remove index
original_dataframe.to_csv('trig.csv', index=False)

#read from a csv
read_dataframe = pd.read_csv('trig.csv')

print(read_dataframe)
```

```python
#access dataframe columns by . notation or ['name']

print(original_dataframe.sin)
print(original_dataframe['sin'])

#to get more than one just do
print(original_dataframe[['sin', 'tan']])

#for rows just use the operations that work for series
#getting the 50th sin entry
print(original_dataframe['sin'].iloc[49])

#also just works direcly
print(original_dataframe.iloc[49])

#to get by index just use .loc
print(original_dataframe.loc[[1, 2, 3], ['sin', 'tan']])
```

```python
#insert a new column into a dataframe
df = pd.DataFrame(
[   [5, 5],
    [6, 7],
    [8, 9],],
    columns= ['a', 'b']
)
df['c'] = [1, 2, 3]
df['d'] = 5
df['e'] = df.a + df.b

#works with boolean operations
df.loc[df.a == 5, 'f'] = 4

print(df)

#to delete columns 
df.drop(columns=['a', 'f'], inplace=True)

print(df)
```

```python
#.apply(), axis=0 works on columns and axis=1 works on rows

df = pd.DataFrame(
[   [5, 5],
    [6, 7],
    [8, 9],],
    columns= ['a', 'b']
)

print(df.apply(func=np.sum, axis=0))
print(df.apply(func=np.sum, axis=1))
```
