```python
import pandas as pd
import numpy as np
```

```python
#Fair Teams Problem, randomly assign coaches to players wwith a 1:3 or 1:4 ratio

coaches = pd.Series(['Aaron', 'Donald', 'Joshua', 'Peter', 'Scott', 'Stephen'], dtype='string')
players = pd.Series(['Asher', 'Connor', 'Elizabeth', 'Emily', 'Ethan', 'Hannah', 'Isabella', 'Isaiah', 'James',
                     'Joshua', 'Julian', 'Layla', 'Leo', 'Madison', 'Mia', 'Oliver', 'Ryan', 'Scarlett', 'William',
                     'Wyatt'], dtype='string')
```

```python
#randomize the series
coaches = coaches.sample(frac=1)
players = players.sample(frac=1)
```

```python
#assign coaches to players by re-indexing the players series with coaches
#make an np array with the coaches, use fancy indexing to make a larger array to be used as an index 

players.index = coaches.to_numpy()[np.arange(len(players)) % len(coaches)]
```
