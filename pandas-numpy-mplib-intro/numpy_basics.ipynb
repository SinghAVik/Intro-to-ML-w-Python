```python
import numpy as np
```

```python
#initalize array
a = np.array([1, 2, 3])
print(a)

b = np.array([[1, 2, 3, 4], [5, 6, 7, 8]], dtype=float)
print(b)
```

```python
#dimension of the array
print(a.ndim)
print(b.ndim)

#shape of the array (row x column)
print(a.shape)
print(b.shape)

#type
print(a.dtype)
print(b.dtype)

#size
print(a.itemsize)
print(b.itemsize)

#total size
print(a.nbytes)
print(b.nbytes)
```

```python
#array manipulation

#get specific element [row, column]
print(b[1,3])

#get specific row
print(b[1,:])

#specific column
print(b[:,3])

#[start:end:step]
print(b[0, 1:4:2])
```

```python
#all 0's
np.zeros((2,3))
```

```python
#all 1's
np.ones((2, 3, 5))
```

```python
#identity matrix
np.identity(3)
```
