```python
import pandas as pd
import numpy as np
```

```python
#discard rows where more than half the values are NaN, elsewhere impute NaNs with the average value per column unless the column is non-numeric using the mode

potholes = pd.DataFrame({
    'length':[5.1, np.nan, 6.2, 4.3, 6.0, 5.1, 6.5, 4.3, np.nan, np.nan],
    'width':[2.8, 5.8, 6.5, 6.1, 5.8, np.nan, 6.3, 6.1, 5.4, 5.0],
    'depth':[2.6, np.nan, 4.2, 0.8, 2.6, np.nan, 3.9, 4.8, 4.0, np.nan],
    'location':pd.Series(['center', 'north edge', np.nan, 'center', 'north edge', 'center', 'west edge',
                          'west edge', np.nan, np.nan], dtype='string')
})
print(potholes)
```

```python
potholes = potholes.dropna(thresh=2)

print(potholes)
```

```python
numeric = potholes.select_dtypes(include='number').columns
str = potholes.select_dtypes(include='string').columns
```

```python
potholes[numeric] = potholes[numeric].fillna(potholes[numeric].mean())
potholes[str] = potholes[str].fillna(potholes[str].mode().iloc[0])
```

```python
print(potholes)
```
