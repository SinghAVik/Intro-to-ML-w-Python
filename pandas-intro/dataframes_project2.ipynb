```python
import pandas as pd
import numpy as np
```

```python
#determine the number of machines carrying each product and how many of them are out of stock
machine_products = pd.DataFrame({
    'machine': ['abc', 'abc', 'def', 'def', 'def', 'ghi'],
    'product': ['skittles', 'soap', 'soap', 'm&ms', 'skittles', 'm&ms'],
    'stock': [10, 0, 15, 2, 0, 3]
})
```

```python
product_numbers = machine_products.groupby('product').agg(machines = ('stock', 'count'), out_of_stock = ('stock', lambda x: (x == 0).sum()))

print(product_numbers)
```
