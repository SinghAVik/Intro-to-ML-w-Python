```python
import pandas as pd
import numpy as np
```

```python
#Ways to make series
series_intigers = pd.Series([1, 2, 3, 4, 5])
series_dictonary = pd.Series({'a': 1, 'b': 2, 'c': 3})
series_strings = pd.Series(['hi', 'hello', 'shalom'], dtype = 'string')
series_narry = pd.Series(np.array([1, 2, 3, 4, 5]))
series_function = pd.Series(np.sin(np.linspace(-np.pi, np.pi, 100)))


print(series_intigers)
print(series_dictonary)
print(series_strings)
print(series_narry)
print(series_function)
```

```python
#use .iloc for positional indexing
print(series_intigers.iloc[0])
print(series_dictonary.iloc[1])
print(series_strings.iloc[1])
print(series_narry.iloc[3])
print(series_function.iloc[73])

#use .loc for label indexing
print(series_dictonary.loc['b'])
```

```python
#rewrite vlaues with this function
series_strings.iloc[2] = 'hola'
series_dictonary.loc['b'] = 99

print(series_strings)
print(series_dictonary)

#to keep index and replace values
array = np.array([5, 6, 7])
series_dictonary.iloc[:] = array

print(series_dictonary)
```

```python
#series operations

#operations are telegraphed to the whole series
series_intigers = series_intigers + 1
print(series_intigers)

#operations between series works via indexing
series_intigers += pd.Series([1, 2, 3, 4, 5], index=[4, 3, 2, 1, 0])
print(series_intigers)

#when adding series with diffrent indexing the new series reverts to sequential indexing
series1 = pd.Series([1, 2, 3, 4, 5], index=[4, 3, 2, 1, 0])
series2 = pd.Series([1, 2, 3, 4, 5])
addpd = series1 + series2 #same as series1.to_numpy() + series2.to_numpy()
print(addpd)

#a series retains indexing when it is added with a np array
addnp = series1 + series1.to_numpy()
print(addnp)

#to add only the index values that are correspoinding to the other series (deletes other values)
series1 = pd.Series([1, 2, 3, 4])
series2 = pd.Series([6, 7], index=[1,3])
series1.loc[series2.index] += series2

print(series1)
```

```python
#use booleans to index series with lables using loc & position using .to_numpy()
series1 = pd.Series([10, 20, 30, 40, 50, 60])
sub30 = series1 < 30

print(sub30)
print(series1.loc[sub30])

#change indexing but label stays the same
series1.index = [5, 2, 3, 4, 1, 6]
sub30 = series1 < 30

print(sub30)
print(series1.loc[sub30])
print(series1.loc[sub30.to_numpy()])

#change indexing of sub30 and there is a dffrence between positional and label indexing
series1 = pd.Series([10, 20, 30, 40, 50, 60])
sub30 = series1 < 30
sub30.index = [5, 2, 3, 4, 1, 6]

print(series1.loc[sub30.to_numpy()])
```

```python
#Boolean indexing when labels and order match
series1 = pd.Series([10, 20, 30, 40, 50, 60])
sub30 = series1 < 30

print(sub30)
print(series1.loc[sub30])

#Part 2: Change the Series' index, then recompute the boolean mask

#Change only the labels, not the positions
series1.index = [5, 2, 3, 4, 1, 6]
sub30 = series1 < 30
print(sub30)

print(series1.loc[sub30]) #Still label-based; selects labels 5 and 2
print(series1.loc[sub30.to_numpy()]) #NumPy mask → no labels → positional selection

#Demonstrate the mismatch that creates label-vs-position differences
series1 = pd.Series([10, 20, 30, 40, 50, 60])
sub30 = series1 < 30

sub30.index = [5, 2, 3, 4, 1, 6]

print(series1.loc[sub30.to_numpy()])  #Works because NumPy array → positional indexing
```

```python
#use boolean operations to combine series
series1 = pd.Series([1, 2, 3, 4, 1, 2, 3, 4], index=['a', 'a', 'a', 'b', 'b', 'b', 'b', 'b'])

a_list   = series1.index == 'a'
sub3    = series1 < 3
grater2    = series1 > 2

and_result = series1[(a_list) & (sub3)]
print(and_result)

or_result = series1[(a_list) | (grater2)]
print(or_result)


#use loc
print(series1.loc[a_list & sub3])
print(series1.loc[a_list | grater2])
```

```python
#pandas has the boolean, Int64, Float64, string data types that are compatible with pd.Na (empty value dataype)
string = pd.Series([True, pd.NA, False], dtype="boolean")

print(string)
print(pd.Series([1, pd.NA, 3], dtype="Int64"))
print(pd.Series([2.0, pd.NA, 7.8], dtype="Float64"))
print(pd.Series(['hi', pd.NA, 'hello'], dtype="string"))

#pd.isna and pd.notna make a boolean series with info
print(pd.isna(string))
print(pd.notna(string))

#when indexing with series that have Na, make sure they are in the correct dtype
ints = pd.Series([1, 2, 3], dtype="Int64")

print(ints.loc[string])
```

```python
#series operations use vectorization to speed up operations
numbers = series_function = pd.Series(np.sin(np.linspace(-np.pi, np.pi, 100)))

print(numbers.mean())

#in certain cases you need to implement your own fuction and to get it to work use apply() to get it on every element
def add_or_multiply(x):
    return x + 1 if x > 0.3 else x * 2

numbers.apply(add_or_multiply)

print(numbers.apply(add_or_multiply)) #you can add trailing params after the func name with ,a,b,c...
```

```python
#some operations return a pointer to a diffrent series while other operations make a copy
#most loc operations return a copy
#use the .copy method to make sure that the returned series is a copy
numbers = series_function = pd.Series(np.sin(np.linspace(-np.pi, np.pi, 100)))
new_numbers = numbers.copy().loc[:70]

print(new_numbers)

#to achive the opposite effect use .replace(Inplace = true)
numbers.loc[:70] = numbers.loc[:70].replace(0, 999)

print(numbers)
```
