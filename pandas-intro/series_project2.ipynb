```python
import pandas as pd
import numpy as np
```

```python
#Car Shopping Problem, determine for which cars is the asking price better than the fair price

asking_prices = pd.Series([5000, 7600, 9000, 8500, 7000], index=['civic', 'civic', 'camry', 'mustang', 'mustang'])
fair_prices = pd.Series([5500, 7500, 7500], index=['civic', 'mustang', 'camry'])
```

```python
good_deal = np.empty(len(asking_prices), dtype=object)

for i in range(len(asking_prices)):
    if asking_prices.iloc[i] > fair_prices.loc[asking_prices.index[i]]:
        good_deal[i] = '_unfair'
    else:
        good_deal[i] = '_fair'

print(good_deal)
```

```python
#place back into the index

asking_prices.index = asking_prices.index + good_deal

print(asking_prices)
```
